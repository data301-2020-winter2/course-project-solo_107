# Analysis

#### Importing the cleaned data set and required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scripts import project_functions

sns.set_theme(style="ticks",font_scale=1.3)
```

```python
df = project_functions.load_and_process('../data/raw/winemag-data.csv')
df.head()
```

## Question 1: How is the price of wine related to points scored? 

#### First, I will look at a scatterplot to compare price and points:

```python
sns.lmplot(
    data=df,
    x="Points", y="Price",
    height=5,
    order = 3,
    palette = 'rocket_r',
    line_kws={"color":"maroon"}
)
```

#### Now, lets look at a scatter plot for each country:

```python
sns.lmplot(x="Points", y="Price", col="Country", data=df,
           col_wrap=2, height=3,line_kws={"color":"maroon"}, order = 3);
```

### Conclusion:
As Price increases, so does Points. However, this relationship is very weak. As Points increases, the Price only slightly increases. Therefore, it is not neccessarily true that all high quality wines are more expensive. Some countries, such as France, Italy, and Austrailia have a higher correlation.

## Question 2: What country produces the best quality wine?

##### First, we need to take a look at the countries that have the highest average score for wine.

```python
points_by_country = pd.DataFrame(df.groupby('Country').mean()['Points'])
points_by_country = points_by_country.sort_values(by = 'Points', ascending = False)[:10].reset_index()
sns.barplot(data = points_by_country, x = 'Country',y='Points',palette='rocket_r').set_title("Average Score of Wine by Country")
```

##### Now we will look at which country has the most perfect scoring wines.

```python
PS = df[ df['Points'].max() == df['Points'] ]
# These are wines that get a maximum score of 100

sns.countplot(data=PS, 
              x='Country',
              order= PS['Country'].value_counts().index[:6],palette='rocket_r'
             ).set_title("Perfect Score Wine by Country (Score = 100)")
```

##### Now we will look at which country scores highest out of the top 5 scoring wines for each country.

```python
points_by_variety = pd.DataFrame(df.groupby('Variety').mean()['Points'])
top10_wines = points_by_variety.sort_values(by = 'Points', ascending = False)[:10].reset_index()['Variety'].to_list()
df_top10 = df[df['Variety'].isin(top10_wines)]
pointsPivot = df_top10.pivot_table(index=['Country','Variety'],values=['Points'],aggfunc='mean')
```

```python
heat_map1 = pointsPivot.pivot_table(index='Variety', columns='Country', values='Points')
# You can separate data with lines
sns.heatmap(heat_map1, cmap='Reds', linecolor='white', linewidth=1)
```

#### Lets look at the overall distributions of the Points data by Country:

```python
sns.set_theme(style="ticks")
f, ax = plt.subplots(figsize=(10, 7))

sns.boxplot(x="Points", y="Country", data=df,
            whis=[0, 100], width=.6, palette="rocket_r")

sns.swarmplot(x="Points", y="Country", data=df,
              size=5, color=".5", linewidth=0,alpha=.3, jitter = 1, dodge= True)

ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
ax.set_title("Points by Country")
```

```python
sns.swarmplot(x="Points", y="Country", data=df)
```

### Conclusion: 
Although all countries have very similar averages of points, it is clear that France overall has the best scoring wine. Although Canada has the highest average, it look as if this could be because of lack of data. France has the most perfect scoring wines and has the highest scoring wines out of the top ten varieties.

## Question 3: What are the top ten most expensive regions for producing wine?

##### Aggregate the dataset into average price by region:

```python
price_by_region = pd.DataFrame(df.groupby('Region').mean()['Price'])
price_by_region = price_by_region.sort_values(by = 'Price', ascending = False)[:10].reset_index()
```

##### Sort the data set and look at the top 10 regions:

```python
plt = sns.barplot(data = price_by_region, 
            x = 'Region', 
            y = 'Price',
            palette = 'rocket_r')
plt.set_title("Top 10 Average Prices by Region")
x = plt.get_xticklabels()
plt.set_xticklabels(x, rotation=90)
```

##### Find the 10 most expensive wines:

```python
wine_by_price = df[['Price','Name','Region']].sort_values(by = 'Price', ascending = False)[:10]
```

##### Which regions produce the top 10 most expensive wines?

```python
plt = sns.barplot(data = wine_by_price, 
            x = 'Region', 
            y = 'Price',
            ci = None,
            palette = 'rocket_r')
plt.set_title("Top 10 Most Expensive Wines by Region")
x = plt.get_xticklabels()
plt.set_xticklabels(x, rotation=45)
```

## Conclusion:
The top ten expensive regions are most probably the ones with the highest average cost for their wines.  Although some regions have high individual wines, it is clear that the regions from the first chart are the most expensive when taking into account all wines in the region. However, if one is looking for a region containing singular highly priced wines, the ten regions in the second chart are more relevant. Overall, since La Romanee appears on both lists, it is fairly safe to say that it is one of the highest end regions.

## Question 4: Which wines are the 'best deal' (high score and low price)?

#### We will look for wines that are of the highest quality. First, lets look at the distribution of points:

```python
sns.histplot(df, x='Points',binwidth=1,stat='count',palette='rocket_r').set_title('Distribution of Points')
```

```python
sns.boxplot(x=df['Points'],palette='rocket_r').set_title('Distribution of Points')
```

#### Lets take at look at all of the wines greater than or equal to 98 points, that are under 100 dollars.

```python
df2 = df[df['Points']>=98]
#df2 = df2[df2['Price']<=500]
df2['Location'] = df2['Province']+', '+df2['Country']
df2
heat_map2 = df2.pivot_table(index='Points', columns='Location', values='Price')
# You can separate data with lines
sns.heatmap(heat_map2, cmap='Reds', linecolor='white', linewidth=1)
```

```python
sns.relplot(
    data=df2,
    x="Points", y="Price",
    hue="Country", 
    kind="line", 
    height=5, aspect=1.5, facet_kws=dict(sharex=False),
    palette = "rocket_r"
)
```

```python
wines = df2.pivot("Location", "Points", "Price")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(wines, annot=True, fmt="d", linewidths=.5, ax=ax)
```
